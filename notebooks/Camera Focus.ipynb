```python
def extract_tete_frames(stage_id):
    PATH = '/Users/samuelalbanie/aims_course/project_two/code/DVD/deinterlaced/2014'
    SRC_PATH = PATH + '/second_interval_frames/' + str(stage_id)
    DEST_PATH = PATH + '/labeled_frames/' + str(stage_id) + '/'
    camera_focus = CameraFocus(src_dir=SRC_PATH)
    log = camera_focus.get_camera_states()
    tete_indices = [i for i, x in enumerate(log) if x == 1]
    chase_indices = [i for i, x in enumerate(log) if x == 2]
    print('Stage id:', str(stage_id))
    print('Number of tete indices: ', len(tete_indices))
    print('Number of chase indices: ', len(chase_indices))
    print('==============================================')
    camera_focus.visualize(lw=5, alpha=0.5, color='green')
    camera_focus.extract_tete_snapshots(DEST_PATH)
```
